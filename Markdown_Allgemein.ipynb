<a id="Sonderzeichen">

#### Sonderzeichen </a> 

Da viele Zeichen in Markdown eine Bedeutung haben und vom Interpreter als befehl gewertet werden, muss dem Interpreter erst erklärt werden dass man ein sonderzeichen schreiben möchte dafür wird ein "\" genutzt, und Vor das Sonderzeichen gestellt. Diesen Vorgang bezeichnet man als "freistellen"   
Solche Sonderzeichen sind zum Beispiel \# \$ \% \\ es gibt aber noch mehr.



<a id="Überschriften">

#### Überschriften </a> 

Ihr könnt bis zu 6 Überschriften verschachteln. dabei wird jede weitere Verschachtelung mit einem zusätzlichen \# angefangen.  
hier ein Beispiel:

\# Das ist die oberste Überschrift  
\## die zweite Überschrift  
\### die dritte Überschrift  
\#### die Vierte Überschrift  
\## und dann wieder eine zweite Überschrift  
\# Hier könnte ein neues Kapitel beginnen mit einer obersten Überschrift   

und so sieht das gerendert aus:

# Das ist die oberste Überschrift 

## die zweite Überschrift   

### die dritte Überschrift

### ToDo-Listen 
</a> 

Abhackbare listen mit \[Leerzeichen] machen in etwa so (mit 'X' im kasten kann man einen Haken setzen):    
\- \[ ] Punkt 1    
\- \[ ] Punkt 2    
\- \[X] Punkt 3    
\- \[ ] Punkt 4

- [ ] Punkt 1  
- [ ] Punkt 2  
- [X] Punkt 3  
- [ ] Punkt 4

</a>

\|Zeile 1| Zeile 1| anderes gedöns|...|

<a id="Textstellen_hervorheben">
 
### Textstellen hervorheben </a> 

- \*\*Text...\*\* Beispiel: **Lorem Ipsum**    
-  \* Text * Beispiel: *Lorem Ipsum*    
- \~\~Text~~ Beispiel ~~Lorem Ipsum~~     
- Farbieger Text:
  - \<span style="color:'farbe'"> Text \</span>  Beispiel: <span style="color:red"> Lorem Ipsum </span>    
  - \<font color="red"> Lorem Ipsum \</font> Beispiel: <font color="red"> Lorem Ipsum </font>
  - \<p style=color:Blue> Lorem Ipsum \</p> Beispiel: <p style=color:Blue> Lorem Ipsum </p>
- Schriftgröße:
    - \<font size=1> Lorem Ipsum \</font> Beispiel:<font size=1> Lorem Ipsum </font>
    - \<font size=3> Lorem Ipsum \</font> Beispiel: <font size=3> Lorem Ipsum </font>
    - Die größte Schriftgröße ist 7
- Schrift Art:
    - \<span style="font-family:Courier New"> Lorem Ipsum \<span> Beispiel: <span style="font-family:Courier New"> Lorem Ipsum </span>
    - \<span style="font-family:Times New Roman"> Lorem Ipsum \<span> Beispiel: <span style="font-family:Times New Roman"> Lorem Ipsum </span>

<a id="Links_md">

## Links
</a>

Links können entweder auf etwas innerhalb des Verzeichnisses oder auf etwas im Internet verlinken.
Ausserdem ist es möglich einen sog. "Tooltip" anzuzeigen also ein Text der Angezeigt wird wenn man mit dem Cursor über dem Link schwebt.
Die Syntax ist in beidem fällen fast gleich:  

Bei einem Link zu einem Verzeichniss wird der Pfad zu der Datei hinterlegt:  
\[Link Text](Pfad_zu_Datei/Datei.pdf "Text des Tooltips")

Bei einem Link zu einer Website wird die URL der Webseite angegeben:  
\[Link Text](https://URL-der-Website.org "Text des Tooltips")

<a id="Bilder_md">
    
## Bilder Einfügen:
</a>

Die Syntax zum einfügen eines Bildes ist relativ einfach und ist an die eines Links angelehnt.

In die \[] kommen statt der Text des Links eine Beschreibung des Bildes (Wird bei nutzung eines Screenreaders genutzt).   
In die \() kommt wie beim Link der Path zur Datei rein oder die URL zum Bild

Angeführt wird das Ganze von einem \! dadurch wird Markdown gezeigt das es den Inhalt darstellen soll.

Beispiel:

\!\[Beschreibung des Bildes](Pfad oder URL zur Datei "Text des Tooltips")
