

## Einführung

Markdown wurde entwickelt um möglichst gut leserlich zu sein, sowohl in gerenderter Version als auch in Source Code.

Da Markdown auf HTML basiert können auch Html-Tags verwendet werden.
Dies bietet uns die einzigartige Möglichkeit das Aussehen unseres Notebooks noch mehr zu personaliesieren.
Das ist allerdings mehr etwas für die Fortgeschrittenen deswegen hier erst mal die Grundlagen.
<br>

## Grundlagen

Allgemein:   
HTML oder Hypertext Markup Language ist einer der Grundsteine nahezu aller Webseiten.   
HTML gibt einer Website ihren Inhalt. <span style='color:red'> NUR </span> den Inhalt. Wenn ihr das ausprobieren wollt, könnt ihr einfach eine Textdatei mit der Endung .html versehen und mit einem Webbrowser öffnen. Der beinhaltete html-Code - den Ihr natürlich reingeschrieben habt- wird euch dann angezeigt.  

Leider sieht das noch nicht so toll aus. Das liegt daran das euer Browser nicht weiß wie er die Sachen darstellen soll.    
Um eueren Browser (bzw. Jupyter) zusagen wie die Sachen dargestellt werden sollen benutzt ihr CSS, dazu später mehr.   

Ein html Befehl wird immer mit einem \<befehl> eingeleitet und mit einem \</befehl> beendet. Dann wird der Befehl auf alle Wörter oder Sachen zwischen dem Beginn und dem Ende angewandt, es gibt auch Befehle, welche nicht beendet werden müssen diese werden hier gesondert markiert.

<a id="Tags">
<br><br>

## Tags
</a>

Das was ich oben noch als Befehle bezeichnet habe ist eigentlich ein html-Tag. Html-Tags können entweder schon vor definiert sein (zum beispiel \<a>) oder von euch selbst definiert sein. (dazu mehr unter [CSS](#CSS_Cascading_Style_Sheets)).
Wichtig ist in erster linie das ihr wisst das html mit Tags arbeitet und nicht mit Befehlen. Tags markieren nämlich nur den Text den ihr auf eine bestimmte Weise behandeln wollt. Befehle dagegen werden vom Computer (oder vom Browser) ausgeführt. Wie man sich Vorstellen kann gibt es viele verschiedene Tags, deswegen werden hier nur einige wenige genannt, welche in kombination mit Markdown die meisten Nutzungsfälle abdecken.

<a id="Images_und_Links">
<br>

## Images und Links 

</a> 
<a id="Links:">
<br>

### Links: 
</a> 

links: können entweder auf eine Datei auf dem Computer oder auf eine Webseite hinweisen. Die Syntax ist im groben das selbe. 

Links nutzen folgende Syntax:

\<a href="URL oder Pfad zur Datei" title="tooltip Text"> Link Text \</a>

Beispiel:

<a href="https://www.google.com/search?q=Katzenbilder" title="KATZEN!!!!"> Ganz viele tolle Bilder </a>

> <div class=alert-warning>
> <b> Vorsicht:</b> <br>
> Beim Kopieren der URL, aus der Suchleiste kann es sein das Informationen über euren Computer und Cookies etc, mit übertragen werden. <br>
> Diese Informationen sind nicht zwingend gefährlich können aber zur Identifikation genutzt werden. Auch Linkshortener wie  bitly oder google sind nicht zuempfehlen da diese nicht sicher sind.<br>
> Für Veröffentlichungen ausserhalb eines Vertrauten Kreises empfiehlt sich <a href="https://wolf-of-seo.de/was-ist/hyperlink/#" style font=color:"red"> eigene Recherche</a> zur Sicherheit von Hyperlinks.
> </div>

<a id="Bilder">
<br><br>

### Bilder </a> 

Bilder funktionieren etwas anders als links da der Browser nicht eine neue Webseite aufmachen soll sondern etwas in die momentane Webseite aus einer anderen Webseite (oder Datei) einbinden soll.

Beispiel:

\<img src="euer Pfad oder URL">

Mit dieser Syntax können wir ein Bild einfügen.  
so können wir mit html neben "nur" einem Bild auch das Bild bis zu einem gewissen Grad verändern, dazu haben wir viele möglichkeiten. Befehle die wir nach Belieben hinzufügen können beinhalten unter anderem:

|Befehl|Auswirkung|
|------|----------|
|alt=""|gibt anweisungen was ein Screenreader ausgeben soll|
|align=""|verändert die Position des Bildes (right middle left)|
|width=""| verändert die breite an, behält das verhältniss bei <br>wenn nicht zusätzlich mit "height" verwendet|
|height=""|verändert die höhe an, behält das Verhältniss bei <br>wenn nicht zusätzlich mit "width" verwendet|
|title=""|gibt einen tooltip an (wird beim drüber hovern sichtbar)|
|id=""|[siehe unten](#ID)|

Hinweis:
Beim verwenden von "align" wird das Bild vom Text "Umflossen". Um das zu verhindern müssen genügend Zeilen Umbrüche eingefügt werden um das Bild freizustellen.

Beispiele:
    
\<img  align="left" width="200" height="100" src="../Bilder/Mdown.jpg"> <br>

<img  align="left" width="200" height="100" src="Bilder/Mdown.jpg">  <br>
<br>
<br>
<br>

\<img src="Bilder/Mdown.jpg" alt="Markdown" style="width:100px;height:60px;"><br>
<img src="Bilder/Mdown.jpg" alt="Markdown" style="width:100px;height:60px;"><br>
<br>
<br>
<br>
\<img  align="right" width="200" height="100" src="Bilder/Mdown.jpg">  <br>
<img  align="right" width="200" height="100" src="Bilder/Mdown.jpg"><br>  

<br>
<br>
<br>
<br>


\<img width="200" height="100" src="Bilder/Mdown.jpg">  <br>
<img width="200" height="100" href="https://denshub.com/de/hugo-post-insert-image/" src="Bilder/Mdown.jpg"><br>  


\<center><br>
\<img src="Bilder/Mdown.jpg" id="Markdown_Bild">  
\</center>
<center>
<img src="Bilder/Mdown.jpg" id="Markdown_Bild"></center>  

<br>
    
\<a href="https://denshub.com/de/hugo-post-insert-image/" title="Mehr zu Bildern"><br>
\<img src="Bilder/Mdown.jpg" id="Markdown_Bild"> \<a>
<br>
<a href="https://denshub.com/de/hugo-post-insert-image/" title="Mehr zu Bildern">
<img src="Bilder/Mdown.jpg" id="Markdown_Bild"></a>  




Auch ist es möglich Bilder in anderen Tags einzuschließen. Zum Beispiel könnnen Bilder in einen \<figure> Tag eingebettet werden. Dies Ermöglicht uns eine \<figcaption> einzufügen also einen Titel bzw einen Untertitel.


Gerade im Fall von Bildern muss die URL nicht zwingend eine Link ins Internet sein es kann sich auch um den Pfad zu einem Bild auf den Computer handeln.

<a id="Text_align">
<br><br>

### Text ausrichten
</a>

Wir kennen ja den "align"-tag von den Bildern. Diesen Tag können wir auch auf Text anwenden.
<br>

\<div style="text-align: right">  
Lorem Ipsum ...  
\</div>
<br>
<div style="text-align: right">  
Lorem ipsum dolor sit amet, consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut labore et dolore magna aliquyam erat, sed diam voluptua. <br>  
At vero eos et accusam et justo duo dolores et ea rebum. Stet clita kasd gubergren, no sea takimata sanctus est Lorem ipsum dolor sit amet. Lorem ipsum dolor sit amet, <br>
consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut labore et dolore magna aliquyam erat, sed diam voluptua. At vero eos et accusam et justo duo dolores et ea rebum. Stet clita kasd gubergren, no sea takimata sanctus est Lorem ipsum dolor sit amet.
</div>
<br>
<hr> 
<br>

\<div style="text-align: center">  
Lorem Ipsum ...  
\</div>
<br>
<div style="text-align: center">  
Lorem ipsum dolor sit amet, consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut labore et dolore magna aliquyam erat, sed diam voluptua. <br>
At vero eos et accusam et justo duo dolores et ea rebum. Stet clita kasd gubergren, no sea takimata sanctus est Lorem ipsum dolor sit amet. Lorem ipsum dolor sit amet,<br>   
consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut labore et dolore magna aliquyam erat, sed diam voluptua. At vero eos et accusam et justo duo dolores et ea rebum. Stet clita kasd gubergren, no sea takimata sanctus est Lorem ipsum dolor sit amet.   
</div>
<br> 
<hr>
<br>

\<div style="text-align: justify">  
Lorem Ipsum ...   
\</div>
<br>
<div style="text-align: justify">  
Lorem ipsum dolor sit amet, consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut labore et dolore magna aliquyam erat, sed diam voluptua.  <br>
At vero eos et accusam et justo duo dolores et ea rebum. Stet clita kasd gubergren, no sea takimata sanctus est Lorem ipsum dolor sit amet. Lorem ipsum dolor sit amet,<br>   
consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut labore et dolore magna aliquyam erat, sed diam voluptua. At vero eos et accusam et justo duo dolores et ea rebum. Stet clita kasd gubergren, no sea takimata sanctus est Lorem ipsum dolor sit amet.   
</div>


<br>
<br>
<a id="ID">
<br><br>
    
### ID
</a>

Eine ID (Identifikation) wird benötigt wenn ihr euch auf etwas beziehen wollt. Das kann ein Bild, eine Formel (siehe [Mathe Allgemein]) oder irgendwas anderes sein.
Da wir uns mit Markdown beschäftigen spreche ich im folgenden von Labels da auch Markdown über die Möglichkeit verfügt Labels einzugügen.


zur Syntax:
Die Syntax ähnelt ein wenig der des Links.

\<a id="das hier ist ein label"> hier das was Ihr Referenziert\</a>

Wenn Ihr euch auf euer Label Bezieht könnt ihr das mit der schon bekannten Link syntax machen.

>WICHTIG:
>Euer Label sollte keine Leerzeichen binhalten, Unterstriche und Bindestriche sind aber in Ordnung. 
>

\[Der Link text](#euer-label)

oder:
\<a href="#euer-label"> Der Link text </a>


hier ein Beispiel:

[dieser Link führt zu einem Bild](#Markdown_Bild)

<a href="#Markdown_Bild"> Dieser Link führt zum selben Bild </a>
<br><br>

<a id="CSS_Cascading_Style_Sheets">
<br><br>
    
## CSS (Cascading Style Sheets)
</a>

>![Important]
>CSS ist mehr etwas für Fortgeschrittene, hier handelt es sich mehr um eine (sehr) kurze Einführung, und weniger wie es "richtig" gemacht wird. 
>

Um euren Html code zu markieren könnt ihr in eure tags, den befehl class='Name' einbinden, Ihr ersetzt natürlich 'Name' durch den Namen den eure Klasse haben soll.  
Alternativ könnt ihr auch eigene Tags erstellen also \<Euer_Tag_Name> ... \</Euer_Tag_Nahme>

Mit euerer class könnt ihr dann über CSS in einem seperaten File einstellen wie eure class dargestellt werden soll.
Dies könnte zum beispiel so aussehen:

```CSS

<style>  
    .className{  
    background-color: tomato;  
    color: white;  
    padding: 10px;  
    }  
</style>   

```

\<style> Umschliest alle euere CSS Definitionen damit weiß der Browser dass es sich um CSS handelt (Im Kurzen...)

Dazu noch kurz etwas wie der Computer (genauer der Browser) euer CSS liest:

1. der Browser geht von einigen Grundlagen aus. <br>Deswegen kann der Browser reinen HTML-code darstellen.
2. Dann kommt das CSS von Jupyter (oder Markdown).<br> Das definiert zum Beispiel wie Tabellen aussehen. 
3. Dann kommt euer CSS das ihr einfügen könnt oder nicht.
4. Danach hat der Browser nochmal die möglichkeit Sachen zuverändern<br>(zum Beispiel für die Nutzung einer Bildschirm Lupe)
5. Jetzt kommen die CSS Befehle die in eueren Tags drin stehen. <br> siehe zum beispiel [Bilder]
6. Schluss endlich gibt es noch eine Ebene auf der ein Screenreader arbeiten kann <br>(also auch vom Browser gesteuert)

Dieser Aufbau scheint kompliziert ist er aber eigentlich gar nicht mal so sehr. Wichtig ist vor allem: 
Jeder schritt kann die Schritte davor Überschreiben. Schritt 3 Überschreibt also alle anweisungen von Schritt 1 und 2 welche in Konflikt stehen.

Ihr Könntet also zum Beispiel eine CSS klasse für Tabellen erstellen und ihr würdet die Darstellung aller Tabellen verändern ohne das ihr die in die Config von eurem Editor reingehen müsst (nicht zu empfehlen).

Um jetzt euer CSS script in euer Notebook einzufügen braucht ihr ein bischen python ---> ihr braucht also am Anfang eures Notebooks eine codezelle:

```python
# Styling notebook
from IPython.core.display import HTML
def css_styling():
    styles = open("./styles/custom.css", "r").read()
    return HTML(styles)
css_styling()
```
