Jetzt ist so eine Gleichung zwar ganz nett bringt uns aber nichts wenn wir damit nicht rechnen k√∂nnen. Daf√ºr bietet uns Sympy verschiedene Optionen:

- ```evalf```: evaluiert die Funktion f√ºr einen bestimmten Punkt

- ```simplify```: vereinfacht eine Funktion
- ```expand```: "klammert aus"

- ```cancel```: "k√ºrzt" Br√ºche
- ```apart```: erzeugt eine Partialbruchzerlegung

Dies sind nur einige wenige von den Funktionen die Sympy anbietet eine etwas [vollst√§ndigere Liste findet ihr in der Dokumentation](https://docs.sympy.org/latest/tutorials/intro-tutorial/simplification.html) und f√ºr mehr Informationen zum [evaluieren einer Funktion finder ihr hier.](https://docs.sympy.org/latest/tutorials/intro-tutorial/basic_operations.html)

Hier einige Beispiele:

``````python
%reset -f
import sympy as smp

x= smp.symbols('x')

y = (20*x**3 + 10*x**2 - 5*x - 15)/(5*x**2 + 10*x + 5)

print('Die Ursprungsdunktion:')
display(y)

# Evaluieren der Funktion "y" am Punkt x=3
print('Funktionswert am Punkt x=3:')
display(y.evalf(subs={x:3}))

# Vereinfachen der Funktion "y"
print('Vereinfachen der Funktion "y":')
display(smp.simplify(y))

# Ausklammern der Gleichung
print('Ausklammern der Funktion "y":')
y1 = smp.expand(y)
display(y1)

# Ausmultiplizieren der Gleichung
print('Ausmultiplizieren von der Gleichung oben:')
display(smp.factor(y1))

#K√ºrzen der Br√ºche (Falls m√∂glich)
print('K√ºrzen der Gleichung "y":')
display(smp.cancel(y))

# Eine Partialbruch zerlegung:
print('Partialbruchzerlegung der Gleichung "y":')
display(smp.apart(y))

# Nullstellen der Gleichung y
print('Nullstellen lauten:')
display(smp.solve(y,x))
``````

#### Differenziale
</a>

Sympy kann Differenziale erstellen, und berechnen. Dies wird zum einen ben√∂tigt um Differentialgleichungen zu l√∂sen und zum anderen um Differenziale zu l√∂sen.
Es gibt verschiedene Arten differenziale zu definieren wir besch√§ftigen uns aber erst mal nur mit der ```diff()``` funktion.

Die ```diff()``` Funktion kann entweder eine Gleichung einfach Differenzieren oder in einem Schritt mehrfach Differenzieren. 
Auch kann ```diff()``` eine Funktion gleich evaluieren.


Durch Konvention werden ableitungen in variablen mit dem selben Zeichen gespeichert mit einem zus√§tzlichen p f√ºr jede Ableitung

y = Gleichung  
y' = y prime = yp = 1. Ableitung  
y''= y prime prime = ypp = 2. Ableitung  
.  
.  
.  

Mehr zur Differenziation findet ihr [in der Dokumentation](https://docs.sympy.org/latest/modules/core.html#sympy.core.function.diff)

``````python
%reset -f
import sympy as smp

x, z = smp.symbols('x z')


y = x**5 + x**4 + x**2 + 25 + smp.sin(x)
display(y)

#1. Ableitung nach x
yp = smp.diff(y,x)
display(yp)

#2. Ableitungen nach x
ypp = smp.diff(y, x, x)
ypp = smp.diff(y, (x, 2))
ypp = smp.diff(yp, x)
display(ypp)

# 3. Ableitung nach x
yppp = smp.diff(y, x, x, x)
yppp = smp.diff(y, (x,3))
yppp = smp.diff(ypp, x)
display(yppp)

# Neue Gleichung in 2 Dimensionen

y = x**5 + (x+z)**8 + z**2 + 25 + smp.sin(z) * smp.cos(x)
display(y)

# 1. Ableitung nach z

ypz = smp.diff(y, z)
display(ypz)

# 3. Ableitung nach z und 2. Ableitung nach x 

ypppzppx = smp.diff(y, (z,3),(x,2))
display(ypppzppx)
``````

#### L√∂sen von Differenzial Gleichungen
</a>

Zum L√∂sen von DGL gibt uns Sympy einige besondere Variablen:

- Eq( ... ), In Eq wird die Differenzial gleichung gespeichert
- func[ ... ], in func werden verschiedene Gleichungen gespeichert welche die DGL ausmachen. Wenn nicht angegeben sucht sich sympy die gleichungen selber.

Zum L√∂sen von DGL`S gibt es in Sympy zwei methoden:

1. Das Symbolische L√∂sen der Gleichung (das wird hier behandelt)
2. Das Numerische L√∂sen der Gleichung mit Scipy

Sympy kann DGLs mit verschiedenen Ans√§tzen l√∂sen. Wir behandeln hier die solve() Methode bei dieser Methode nutzt Sympy automatisch die "Beste L√∂sung" da die Heuristik welche L√∂sungsmethode die beste ist aber nicht perfekt ist kann es sinnvoll sein andere Methoden vorzuschreiben daf√ºr einfach in die [Dokumentation](https://docs.sympy.org/latest/guides/solving/solve-ode.html#solve-an-ordinary-differential-equation-ode) schauen.

In diesem Beispiel l√∂sen wir eine Differenzialgleichung einfache relativ einfache Differenzialgleichung.


```{sidebar} Die Gleichung:
$
\begin{align*}
0 = \ddot{x} + x \\
\end{align*}
$
```

Bei dem Beispiel f√§llt auf das wir x als eine Funtktion Definieren und nicht als Symbol.   
Das ist wichtig da wir davon ausgehen das in unserer DGL x von t (also der Zeit) abh√§ngig ist.

`````{admonition} Tipp:
:class: tip, dropdown
Es gibt eine M√∂glichkeit die Funktion 'x' von anfang an von t abh√§nhig zumachen:
```python
x = smp.Function('x')(t)
```
Dies hat leider den Nachteil das ```smp.Eq``` das nicht gut findet wenn wir dann damit rechnen (vor allem wenn wir Anfangsbedinungen festlegen wollen).   
Deswegen stellen wir diese Abh√§ngigkeit erst in der ```Equation``` her. Aus diesem Grund tauchen auf einmal Klammern beim x auf.   
Deswegen wenn ihr irgendwelche Fehlermeldungen bekommt die ihr nicht versteht schaut einfach mal ob ihr das Umsetzen k√∂nnt...
`````


``````python
%reset -f
import sympy as smp

t= smp.symbols('t', real=True)

x = smp.Function('x')

equation = smp.Eq(x(t).diff(t,t) + x(t), 0)

l√∂sung = smp.dsolve(equation)

display(equation)
display(l√∂sung)
``````

<br>
Wir bekommen eine Gleichung f√ºr x welche aber noch von den Integrations-Konstanten $C_1$ und $C_2$ abh√§ngig sind. Da wir das nicht zwingend wollen, k√∂nnen wir in ```dsolve()``` ics angeben.    

F√ºr unsere Gleichung gehen wir einfach mal von folgenden Anfangsbedingungen aus:

1.  Am anfang haben wir die Auslenkung $x(t=0) = x_0$
2.  Am anfang haben wir die Geschwindigkeit $\dot{x}= v_0$

``````python
%reset -f
import sympy as smp

t= smp.symbols('t', real=True)

"""
Anfangsbadingungen:
x(t=0) = 10 m
xp(t=0) = 5 m/s
"""

x = smp.Function('x')

equation = smp.Eq(x(t).diff(t,t) + x(t), 0)

l√∂sung = smp.dsolve(equation, ics = {x(0): 10, smp.diff(x(t), t).subs(t,0): 5})

display(equation)
display(l√∂sung)
``````
