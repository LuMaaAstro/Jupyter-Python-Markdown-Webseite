## Mathe Formel<a id="Mathe_Formel"> </a> 

### Mathe bereich:<a id="Mathe_bereich"> </a> 
der Mathebereich wird gebraucht um Markdown zuzeigen was eine Formel ist und was einfach nur seltsamer text.    
der Mathebereich kann entweder entkoppelt vom text oder im text bleiben,   
wenn der Mathebereich im text sein soll müsst begrenzt ihr ihn mit einem Dollarzeichen vorne und hinten.   
hier ein Beispiel Ihr schreibt: 'Dollarzeichen' 2+2=4 'Dollarzeichen' und es kommt raus: $2+2=4$

Wenn die Formel entkoppelt vom Text stehen soll müsst ihr die Formel mit \$\$ begrenzen. Um nicht durcheinander zukommen empfiehlt es sich dafür im Texteditor eine neue Zeile zubeginnen.
zum Beispiel Ihr schreibt:   

'DollarzeichenDollarzeichen'  
    2 + 2 = 4   
'DollarzeichenDollarzeichen'  

und raus kommt: 

$$
 2+2=4
$$

### Matrizen und Vektoren<a id="Matrizen_und_Vektoren"> </a> 

Matrixen sind relativ leicht, Matrizen werden auch im Mathe bereich beschrieben  und werden mit \begin{bmatrix} gestartet und mit \end{matrix} beendet diese Syntax wird in Markdown öfters verwendet. Dabei wird zuerst eine besondere Syntax eingeleitet und später wieder beendet alles zwischen dem Anfang und dem Ende wird anders behandelt als einfacher text.
Die einzelnen einträge innerhalb einer Zeile werden mit & symbol getrennt. Um eine neue Zeile anzufangen wird ein doppelter Backslash genutzt (\\). 
In Natura könnte das ganze in etwa so aussehen:  
<span style="color:orange"> Hinweis: da Markdown alle befehle sonst umsetzt sind die Beispiele nicht zum kopieren geiegnet, falls ihr trotzdem den Code kopieren wollt müsst ihr erst die gerenderte version mit doppelklick beenden und dann den entsprechenden Code kopieren der richtige Code ist meist nach oder unter dem Beispiel, oft nach dem Satz "raus kommt"(ohne gewähr!). 
</span>

/begin{bmatrix}  
11 & 24 & 7 & 20 & 3 \\ \\   
4 & 12 & 25 & 8 & 16 \\ \\  
17 & 5 & 13 & 21 & 9 \\ \\  
10 & 18 & 1 & 14 & 22 \\ \\  
23 & 6 & 19 & 2 & 15 \\ \\  
/end{bmatrix}

Und so sieht es gerendert aus:

\begin{bmatrix}  
11 & 24 & 7 & 20 & 3 \\   
4 & 12 & 25 & 8 & 16 \\  
17 & 5 & 13 & 21 & 9 \\  
10 & 18 & 1 & 14 & 22 \\  
23 & 6 & 19 & 2 & 15 \\  
\end{bmatrix}

Mit diesem Wissen ist es nicht schwer einen Vektor zu defienieren da ein vektor einfach nur eine Matrix ist die in jeder zeile nur einen Wert besitzt

Wenn Formeln untereinander geordnet sein soll verändert sich die Syntax indiesem Fall muss der Mathe bereich mit einem  ohne \$ dafür mit \\begin{align*} begonnen werden und mit \\end{align*} beendet werden
um Markdown klar zumachen welche zeiches untereinander ist wird ein & Symbol genutzt. um eine Neue Zeile zubeginnen muss ein \\ an das Ende jeder Zeile kommen

Bsp:

\\begin{align*}

a & = b \\ \     

X &\sim {\sf Norm}(10, 3) \\ \     

5 & \le 10     

\\end{align*}     

daraus wird:



\begin{align*}
a & = b \\
X &\sim {\sf Norm}(10, 3) \\
5 & \le 10
\end{align*}

NameEurerVariable = ...

```python
Eingabe = ''
Eingabe = input ('Hier bitte irgendwas eingeben und Enter drücken')

if Eingabe != '':
    print('Der nutzer hat eine Eingabe gemacht')
else:
    print ('Keine Eingabe erfolgt')
```

### while-Schleifen <a id="while-Schleifen"> </a> 

Eine while-Schleife in Python ist eine Kontrollstruktur, die verwendet wird, um einen Block von Code so lange auszuführen, wie eine bestimmte Bedingung wahr (True) ist. Die Schleife überprüft zu Beginn jeder Iteration die Bedingung. Solange die Bedingung wahr ist, wird der Codeblock im Schleifenkörper ausgeführt. Sobald die Bedingung falsch wird, wird die Schleife beendet, und die Programmausführung wird nach der Schleife fortgesetzt.

Hier ist eine kurze Erklärung der Hauptkomponenten und Verwendungszwecke von while-Schleifen:

- **Bedingung:** Die Bedingung, die am Anfang jeder Schleifeniteration überprüft wird. Solange diese Bedingung wahr ist, wird die Schleife ausgeführt.

- **Schleifenkörper:** Der Block von Anweisungen, der innerhalb der Schleife ausgeführt wird. Dieser Block kann eine oder mehrere Anweisungen enthalten.

- **Aktualisierung der Bedingung:** Innerhalb des Schleifenkörpers müssen Sie sicherstellen, dass die Bedingung schließlich falsch wird. Andernfalls haben Sie eine Endlosschleife.

while-Schleifen werden oft verwendet, wenn Sie nicht im Voraus wissen, wie oft eine Schleife wiederholt werden soll, sondern dies von einer sich ändernden Bedingung abhängt. Es ist wichtig, sicherzustellen, dass die Bedingung im Laufe der Schleife geändert wird, damit die Schleife schließlich endet und nicht unendlich weiterläuft.

Im folgenden ein Beispiel, wo immer wieder um eine Eingabe gebeten wird bis der User die "geheime" 
Zahl erät. 

```python
geheime_Zahl = 5
Eingabe = 0

bool = False


while bool == False:
    Eingabe = input('bitte geben Sie eine ganze Zahl an:')
    if int(Eingabe) == geheime_Zahl:
        print ('Hurra, genau ' + str(Eingabe) + ' ist die geheime Zahl')
        bool = True    
    else:
        print ('leider ist' + str(Eingabe) + 'nicht die geheime Zahl, probiere es einfach Nochmal')
        
    

```

### try-Schleifen<a id="try-Schleifen"> </a> 

Manchmal weiß man nicht ob der Code so wie er ist tatsächlich funktioniert, dann muss man ihn absichern. Ein Beispiel dafür ist der Code da oben $\uparrow$ 
Wenn jemand so schlau ist und da was eingibt was nicht in einen integer umgewandelt werden kann dann wirft python mit Fehlern um sich.
Falls man das nicht möchte (vorallem bei Systemkritischen Code zum beispiel bei Robotern) kann man sich über eine `try` Schleife absichern, falls ein Fehler auftaucht geht nicht der ganze code kapputt stattdessen wird der Code Schnipsel ausgeführt der am ende mit  except steht. Das klingt kompliziert deswegen ist hier der code von oben nochmal diesmal aber abgesichert.

Noch ein kleiner Hinweis da wir nicht immer 100% wissen welcher Fehler auftritt wir aber der Funktion `exception` einen Wert geben Müssen geben wir ihr einfach `Exception` als wert damit haben wir alle Fehler abgedeckt.

```python
geheime_Zahl = 5
Eingabe = 0

bool = True

while bool:
    try:
        Eingabe = input('bitte geben Sie eine ganze Zahl an:')
        if int(Eingabe) == geheime_Zahl:
            print ('Hurra, genau ' + str(Eingabe) + ' ist die geheime Zahl')
            bool = False    
        else:
            print ('leider ist' + str(Eingabe) + 'nicht die geheime Zahl, probiere es einfach Nochmal')
    except Exception:
        print ('deine Eingabe konnte leider nicht in einen Integer umgewandelt werden das heißt: DU HAST KEINE ZAHL EINGEGEBEN!!!! deswegen probiere es bitte nochmal aber diesmal bitte mit etwas wie 1 oder 5 oder 1516511681138131883317381846713')
    
```

### for-Schleifen <a id="for-Schleifen"> </a> 

Eine for-Schleife in Python ist eine Kontrollstruktur, die verwendet wird, um über eine Sequenz von Elementen zu iterieren, wobei bei jedem Schleifendurchlauf der Codeblock für jedes Element in der Sequenz ausgeführt wird. Hier ist eine kurze Erklärung der Hauptkomponenten und Funktionsweise von for-Schleifen:

- Sequenz: Die Sequenz, über die die Schleife iteriert, kann eine Liste, ein Tupel, eine Zeichenkette, ein Dictionary oder eine andere Sequenz sein.

- Element: In jedem Schleifendurchlauf wird das aktuelle Element der Sequenz ausgewählt und in einer Variable gespeichert, die Sie in der Schleife verwenden können.

- Codeblock: Der Code im Schleifenkörper wird einmal für jedes Element in der Sequenz ausgeführt. Sie können Anweisungen im Schleifenkörper platzieren, um auf das aktuelle Element zuzugreifen und Operationen durchzuführen.

- Iteration: Die for-Schleife führt die Schleife so oft aus, wie es Elemente in der Sequenz gibt, und beendet die Schleife automatisch, wenn alle Elemente verarbeitet wurden.


for-Schleifen sind sehr nützlich, wenn Sie eine wiederholte Aktion für jedes Element in einer Sequenz ausführen müssen. Sie bieten eine effiziente und saubere Möglichkeit, Iterationen in Python durchzuführen.

hier noch ein kurzes Beispiel für for- Schleifen:



```python
fruits = ["Apfel", "Banane", "Kirsche"]
print ('Ohne for Schleife:' )
print (fruits)
print ('mit for-Schleife:')
for fruit in fruits:
    print(fruit)

```

- Interaktive Ereignisse: Tkinter ermöglicht die Verarbeitung von Benutzerereignissen wie Mausklicks und Tastatureingaben. Sie können Funktionen an Ereignisse binden, um auf Benutzerinteraktionen zu reagieren.

- Anpassbarkeit: Sie können das Erscheinungsbild und Verhalten von Tkinter-Widgets anpassen, um die Benutzeroberfläche nach Ihren Wünschen zu gestalten.

- Unterstützung für Dialogfelder: Tkinter bietet vordefinierte Dialogfelder für häufige Aufgaben wie Dateiauswahl, Farbauswahl und Texteingabe.

- Integration mit Python: Tkinter ist Teil der Python-Standardbibliothek und erfordert keine zusätzlichen Installationen. Sie können Tkinter in Ihren Python-Anwendungen sofort verwenden.

Tkinter ist eine großartige Wahl, wenn Sie eine einfache und plattformübergreifende GUI für Ihre Python-Anwendungen erstellen möchten. Es ist besonders nützlich für kleine bis mittelgroße Projekte und bietet eine schnelle Möglichkeit, Benutzeroberflächen zu erstellen, ohne auf externe Bibliotheken oder Frameworks zurückgreifen zu müssen.

Mit Matplotlib können Benutzer eine breite Palette von Diagrammtypen, einschließlich Linien-, Balken- und Streudiagrammen, Histogrammen, 3D-Plots und mehr, erstellen. Die Bibliothek bietet eine hohe Anpassbarkeit und Kontrolle über das Erscheinungsbild und die Eigenschaften der erstellten Diagramme, was es ideal für die wissenschaftliche Datenvisualisierung, Datenanalyse, Präsentationen und Berichterstellung macht.

Die Fähigkeit, sowohl einfache als auch komplexe Diagramme zu erstellen und gleichzeitig die Möglichkeit zur Feinanpassung zu bieten, macht Matplotlib zu einer der am häufigsten verwendeten Bibliotheken für die Erstellung von grafischen Darstellungen in der Python-Programmierung und Datenvisualisierung.

Die Bibliothek ist Open Source, gut dokumentiert und in der Python-Community weit verbreitet. SymPy bietet eine hohe Flexibilität und Kontrolle bei symbolischen Berechnungen und ist ein leistungsfähiges Werkzeug für mathematische Analysen und Modellierungen.
