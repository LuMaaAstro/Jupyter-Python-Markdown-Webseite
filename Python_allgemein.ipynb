<a id="if-Schleifen">

### If-Schleifen  </a> 

If Schleifen dienen dazu den Zustand eines Programs zu überprüfen. Im folgenden Beispiel wird die Variable "Eingabe" erschaffen, und um einen User input gebeten. Sollte der User keinen Input geben und nur Enter drücken wird ein anderer Code ausgewertet als wenn der User irgendwas eingibt.

````python
Eingabe = ''


Eingabe = input ('Hier bitte irgendwas eingeben und Enter drücken')

def f(Eingabe):
    if Eingabe != '':
        print('Der nutzer hat eine Eingabe gemacht')
    else:
        print ('Keine Eingabe erfolgt')

f(Eingabe)
````

Falls die if-schleife nicht ausgeführt wurde wird der Code weiter ausgeführt, falls ihr aber nach der ersten if-Frage gleich noch eine Abfrage machen wollt könnt ihr diese einfach mit einem
`elif` abfragen. Dabei gilt zu beachten das die Abfrage mit elif nur erfolgt wenn die erste Abfrage nicht erfolgreich war.

Falls keine eurer Abfragen ausgelöst wurden kann es sein, dass ihr trotzdem einen Code ausführen wollt: dafür hängt ihr an das Ende eurer Abfragen reihe noch ein `else:` an. 

<a id="while-Schleifen">

### while-Schleifen  </a> 

Eine while-Schleife in Python ist eine Kontrollstruktur, die verwendet wird, um einen Block von Code so lange auszuführen, wie eine bestimmte Bedingung wahr (True) ist. Die Schleife überprüft zu Beginn jeder Iteration die Bedingung. Solange die Bedingung wahr ist, wird der Codeblock im Schleifenkörper ausgeführt. Sobald die Bedingung falsch wird, wird die Schleife beendet, und die Programmausführung wird nach der Schleife fortgesetzt.

Hier ist eine kurze Erklärung der Hauptkomponenten und Verwendungszwecke von while-Schleifen:

- **Bedingung:** Die Bedingung, die am Anfang jeder Schleifeniteration überprüft wird. Solange diese Bedingung wahr ist, wird die Schleife ausgeführt.

- **Schleifenkörper:** Der Block von Anweisungen, der innerhalb der Schleife ausgeführt wird. Dieser Block kann eine oder mehrere Anweisungen enthalten.

- **Aktualisierung der Bedingung:** Innerhalb des Schleifenkörpers müssen Sie sicherstellen, dass die Bedingung schließlich falsch wird. Andernfalls haben Sie eine Endlosschleife (was manchmal gewünscht ist).

while-Schleifen werden oft verwendet, wenn Sie nicht im Voraus wissen, wie oft eine Schleife wiederholt werden soll, sondern dies von einer sich ändernden Bedingung abhängt. Es ist wichtig, sicherzustellen, dass die Bedingung im Laufe der Schleife geändert wird, damit die Schleife schließlich endet und nicht unendlich weiterläuft.

Im folgenden ein Beispiel, wo immer wieder um eine Eingabe gebeten wird bis der User die "geheime" Zahl erät. 

````python
geheime_Zahl = 5
Eingabe = 0

bool = False


while bool == False:
    Eingabe = input('bitte geben Sie eine ganze Zahl an:')
    if int(Eingabe) == geheime_Zahl:
        print ('Hurra, genau ' + str(Eingabe) + ' ist die geheime Zahl')
        bool = True    
    else:
        print ('leider ist' + str(Eingabe) + 'nicht die geheime Zahl, probiere es einfach Nochmal')
        
````

### try-Schleifen </a> 

Manchmal weiß man nicht ob der Code so wie er ist, tatsächlich funktioniert. Dann muss man ihn absichern. Ein Beispiel dafür ist der Code da oben $\uparrow$ 
Wenn jemand so schlau ist und da was eingibt was nicht in einen integer umgewandelt werden kann dann wirft Python mit Fehlern um sich.
Falls man das nicht möchte (vorallem bei Systemkritischen Code zum beispiel bei Robotern) kann man sich über eine `try` Schleife absichern, falls ein Fehler auftaucht geht nicht der ganze code kapputt stattdessen wird der Code Schnipsel ausgeführt der am ende mit except steht. Das klingt kompliziert deswegen ist hier der code von oben nochmal diesmal aber abgesichert.

Noch ein kleiner Hinweis da wir nicht immer 100% wissen welcher Fehler auftritt wir aber der Funktion `exception` einen Wert geben Müssen geben wir ihr einfach `Exception` als wert damit haben wir alle Fehler abgedeckt.

````python
geheime_Zahl = 5
Eingabe = 0

bool = True

while bool:
    try:
        Eingabe = input('bitte geben Sie eine ganze Zahl an:')
        if int(Eingabe) == geheime_Zahl:
            print ('Hurra, genau ' + str(Eingabe) + ' ist die geheime Zahl')
            bool = False    
        else:
            print ('leider ist' + str(Eingabe) + 'nicht die geheime Zahl, probiere es einfach Nochmal')
    except Exception:
        print ('deine Eingabe konnte leider nicht in einen Integer umgewandelt werden das heißt: DU HAST KEINE ZAHL EINGEGEBEN!!!! deswegen probiere es bitte nochmal aber diesmal bitte mit etwas wie 1 oder 5 oder 1516511681138131883317381846713')
    
````

<a id="for-Schleifen">

### for-Schleifen  </a> 

Eine for-Schleife in Python ist eine Kontrollstruktur, die verwendet wird, um über eine Sequenz von Elementen zu iterieren, wobei bei jedem Schleifendurchlauf der Codeblock für jedes Element in der Sequenz ausgeführt wird.   
Hier ist eine kurze Erklärung der Hauptkomponenten und Funktionsweise von for-Schleifen:

- Sequenz: Die Sequenz, über die die Schleife iteriert, kann eine Liste, ein Tupel, eine Zeichenkette, ein Dictionary oder eine andere Sequenz sein.

- Element: In jedem Schleifendurchlauf wird das aktuelle Element der Sequenz ausgewählt und in einer Variable gespeichert, die Sie in der Schleife verwenden können.

- Codeblock: Der Code im Schleifenkörper wird einmal für jedes Element in der Sequenz ausgeführt. Sie können Anweisungen im Schleifenkörper platzieren, um auf das aktuelle Element zuzugreifen und Operationen durchzuführen.

- Iteration: Die for-Schleife führt die Schleife so oft aus, wie es Elemente in der Sequenz gibt, und beendet die Schleife automatisch, wenn alle Elemente verarbeitet wurden.


for-Schleifen sind sehr nützlich, wenn Sie eine wiederholte Aktion für jedes Element in einer Sequenz ausführen müssen. Sie bieten eine effiziente und saubere Möglichkeit, Iterationen in Python durchzuführen.

hier noch ein kurzes Beispiel für for- Schleifen:



````python
fruits = ["Apfel", "Banane", "Kirsche"]
print ('Ohne for Schleife:' )
print (fruits)
print ('mit for-Schleife:')
for fruit in fruits:
    print(fruit)
````

## Funktionen
</a>

Funktionen sind Befehle die ihr selber definieren könnt.

In Funktionen könnt ihr sozusagen kleine Programmschnipsel speichern und aufrufen. Dies ist besonders dann Nützlich wenn ihr immer wieder ein und die Selbe sache ausführen musst. 

Es ist Konvention eigene Funktionen mit einem kleinen Buchstaben zu beginnen für den Code ist das zwar egal, wer aber sich länger mit Programmieren beschäftigt wird noch Klassen kennenlernen, welche mit großem Anfangsbuchstaben gekennzeichnet werden. 

Funktionen existieren erstmal "ausserhalb des Codes" alle Variablen die ihr in der Funktion Definiert sind also erst mal nicht existent für den ganzen rest des Codes. Wenn ihr also irgendwas mit den Daten die ihr habt machen wollt, müsst ihr der Funktion diese Daten erst mal übergeben. Das macht ihr indem ihr in die runden Klammern Variablen reinschreibt. Diese Variablen übernehmen die "Kommunikation" nach aussen. 

Um Daten auch wieder aus der Funktion rauszubekommen könnt ihr entweder eine Variable als global bezeichnen (Unerwünscht und Unpraktisch) oder eine oder mehr Variablen mit ```return``` ausgeben.
 

Funktionen werden wie folgt definiert:

```python
def nameEurerFunktion(Benötigte, Variable):
    #eure Funktion: zum Beispiel zur Berechnung einer Ganzwichtigen Rechnung
    Ausgabe = 2 * Benötigte + Variablen 
    return Ausgabe
```

Erklärung:

Oben wird eine Funktion namens "nameEurerFunktion" definiert. Diese Funktion benötigt die Angaben "Benötigte" und "Variable" das können entweder andere Variablen sein oder irgendwelche Daten.

Innerhalb der Funktion wird dann eine Rechnung mit den Variablen ausgeführt und das Ergebnis in form der Variable "Ausgabe" ausgegeben.

Um diese Funktion jetzt Aufzurufen können wir folgende Syntax verwenden.

```python
a = nameEuererFunktion(Variable1, Variable2)
b = nameEuererFunktion(Variable3, Variable4)
```

##### Kurze Beschreibung

NumPy, kurz für "Numerical Python", ist eine der wichtigsten und weit verbreiteten Python-Bibliotheken für numerische Berechnungen und wissenschaftliches Rechnen. Die Bibliothek bietet leistungsstarke Datenstrukturen und Funktionen, die speziell für das Arbeiten mit großen Datenmengen und numerischen Berechnungen entwickelt wurden.

- Ndarray: Das zentrale Element von NumPy ist das ndarray (N-dimensional Array), das eine effiziente Datenstruktur für das Speichern und Arbeiten mit mehrdimensionalen Arrays darstellt. Diese Arrays können von geringer bis hoher Dimension sein und sind optimiert, um mathematische Operationen auf ganzen Arrays in einem einzigen Schritt auszuführen.

Diese Funktionen sind vektorisiert, was bedeutet, dass sie auf Elemente in einem Array gleichzeitig angewendet werden, was die Effizienz erheblich steigert.

- Broadcasting: NumPy ermöglicht das Durchführen von Operationen auf Arrays mit unterschiedlichen Formen, indem es die sogenannte "Broadcasting"-Regel verwendet. Dies erleichtert das Arbeiten mit Arrays unterschiedlicher Größe und Form.

- Mathematische Funktionen: Die Bibliothek bietet eine umfangreiche Sammlung von mathematischen Funktionen, statistischen Funktionen und linearen Algebra-Operationen, die für wissenschaftliche und numerische Berechnungen von entscheidender Bedeutung sind.

- Datei-Ein- und Ausgabe: NumPy bietet Funktionen zum Lesen und Schreiben von Daten aus und in verschiedene Formate, einschließlich CSV, Textdateien und binäre Dateien.

- Integration mit anderen Bibliotheken: NumPy wird häufig in Kombination mit anderen wissenschaftlichen Python-Bibliotheken wie SciPy, Matplotlib und Pandas verwendet, um umfangreiche Datenanalysen und wissenschaftliche Simulationen durchzuführen.

Durch seine Fähigkeit, effiziente numerische Berechnungen durchzuführen und umfangreiche Tools zur Verfügung zu stellen, hat NumPy einen wesentlichen Beitrag zur Verbreitung von Python im Bereich des wissenschaftlichen Rechnens geleistet und ist ein unverzichtbares Werkzeug für Datenwissenschaftler, Ingenieure und Forscher.

#### SimPy </a>
##### Kurze Beschreibung 
SymPy ist eine Python-Bibliothek für symbolische Mathematik. Sie ermöglicht die symbolische Manipulation von mathematischen Ausdrücken, anstelle von numerischen Berechnungen. Mit SymPy können mathematische Gleichungen, Algebra, Analysis und symbolische Berechnungen durchgeführt werden, ohne auf numerische Näherungen angewiesen zu sein.

SymPy ist nützlich für Aufgaben wie das Lösen von Gleichungen, das Ableiten und Integrieren von Funktionen, das Vereinfachen von mathematischen Ausdrücken, das Erstellen von Symbolen und Variablen, das Arbeiten mit Polynomen und das Durchführen von algebraischen Transformationen. Es ist eine wertvolle Ressource in wissenschaftlichen Berechnungen, Ingenieurwesen, Mathematik und anderen Disziplinen, in denen genaue symbolische Berechnungen erforderlich sind.

Die Bibliothek ist Open Source, gut dokumentiert und in der Python-Community weit verbreitet. SymPy bietet eine hohe Flexibilität und Kontrolle bei symbolischen Berechnungen und ist ein leistungsfähiges Werkzeug für mathematische Analysen und Modellierungen.

Dokumentation: [SimPy-Dokumentation](https://simpy.readthedocs.io/en/latest/)
