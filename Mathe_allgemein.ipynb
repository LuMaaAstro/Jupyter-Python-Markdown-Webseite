
## Mathe bereich:<a id="Mathe_bereich"> </a> 
der Mathebereich wird gebraucht um Markdown zuzeigen was eine Formel ist und was einfach nur seltsamer text.      
der Mathebereich kann entweder entkoppelt vom text oder im text bleiben,   
wenn der Mathebereich im text sein soll müsst begrenzt ihr ihn mit einem Dollarzeichen vorne und hinten.   
hier ein Beispiel Ihr schreibt: und es kommt raus: $2+2=4$  

Wenn die Formel entkoppelt vom Text stehen soll müsst ihr die Formel mit \$\$ begrenzen. Um nicht durcheinander zukommen empfiehlt es sich dafür im Texteditor eine neue Zeile zubeginnen.  
zum Beispiel Ihr schreibt:   

'DollarzeichenDollarzeichen'  
    2 + 2 = 4   
'DollarzeichenDollarzeichen'  

und raus kommt: 

$$
 2+2=4
$$

## Matrizen und Vektoren<a id="Matrizen_und_Vektoren"> </a> 

Matrixen sind relativ leicht, Matrizen werden auch im Mathe bereich beschrieben  und werden mit \\begin{bmatrix} gestartet und mit \end{matrix} beendet.   
Diese Syntax wird in Markdown öfters verwendet. Dabei wird zuerst eine besondere Syntax eingeleitet und später wieder beendet.  
Alles zwischen dem Anfang und dem Ende wird anders behandelt als einfacher text.  
Wenn ihr diese besonderen Art von "Mathe Bereich" nutzen wollt und ihr euer Notebook mit Pandoc in PDF umwandeln wollt, müsst ihr vor Begin und am Ende des Sonderbereichs noch mal zwei \$ setzen.  

Die einzelnen einträge innerhalb einer Zeile werden mit & symbol getrennt. Um eine neue Zeile anzufangen wird ein doppelter Backslash genutzt (\\).    
In Natura könnte das ganze in etwa so aussehen:    


/begin{bmatrix}  <br>
11 & 24 & 7 & 20 & 3 \\ \\ <br>
4 & 12 & 25 & 8 & 16 \\ \\    <br>
17 & 5 & 13 & 21 & 9 \\ \\    <br>
10 & 18 & 1 & 14 & 22 \\ \\    <br>
23 & 6 & 19 & 2 & 15 \\ \\    <br>
/end{bmatrix}<br>

Und so sieht es gerendert aus:

$$
\begin{bmatrix}  
11 & 24 & 7 & 20 & 3 \\   
4 & 12 & 25 & 8 & 16 \\  
17 & 5 & 13 & 21 & 9 \\  
10 & 18 & 1 & 14 & 22 \\  
23 & 6 & 19 & 2 & 15 \\  
\end{bmatrix}
$$

Mit diesem Wissen ist es nicht schwer einen Vektor zu defienieren da ein Vektor einfach nur eine Matrix ist, in der in jeder zeile nur einen Wert besitzt

Neben "bmatrix" gibt es noch andere arten Matrizen darzusetellen:

- matrix $\to$ ohne Klammern
- pmatrix $\to$ (...)
- Bmatrix $\to$ {...}
- vmatrix $\to$ |...|
- Vmatrix $\to$ ||...||


## Klammern

Manchmal möchte man größere Formeln in Klammern setzen. Da kann man entweder einfach Klammern in die Formeln setzen.

Beispiel:(\frac{2x}{5-y}+1)  

$$
(\frac{2x}{5-y}+1)
$$


Leider bleiben die Kammern in diesen Fällen immer gleich groß das sieht natürlich nicht so toll aus. Aus diesem Grund können wir $\LaTeX$ sagen das es die Klammern mit dem Inhalt skalieren soll. 
Dafür geben wir vor die Klammern ein \right oder ein \left (je nachden welche seite der Klammer wir gerade haben)

Beispiel: \left( \frac{2x}{5-y}+1 \right)

$$
\left( \frac{2x}{5-y}+1 \right)
$$


## Formelsysteme ausrichten

Wenn Formeln untereinander geordnet sein soll verändert sich die Syntax indiesem Fall muss der Mathe bereich mit mit \\ begin{align*} begonnen und mit \\ end{align*} beendet werden.  
Um $\LaTeX$ klar zumachen, welche Zeichen untereinander sind wird ein & Symbol genutzt.  
Um eine Neue Zeile zubeginnen muss ein \\ \\ an das Ende jeder Zeile kommen

Bsp:

\$\$
\\begin{align*}
a & = b \\ \    
X &\sim {\sf Norm}(10, 3) \\ \     
5 & \le 10     
\\end{align*}     
\$\$  
daraus wird:


$$
\begin{align*}
a & = b \\
X &\sim {\sf Norm}(10, 3) \\
5 & \le 10
\end{align*}
$$




## Mathe Symbole  <a id="Mathe_Symbole"> </a> 


Die Mathemathischen Symbole müsst ihr immer im Mathe bereich anwenden, ansonsten ist es nur seltsamer Text


|Schreibweise|Umsetzung|  
|-----------:|:--------|   
|Irgendwas^{Zahl oder Text}|$Irgendwas^{Zahl oder Text}$|   
|Irgendwas_{Irgendwas}|$Irgendwas_{Irgendwas}$|   
|<hr>|<hr>|  
|\frac{Zähler}{Nenner}|$\frac{Zähler}{Nenner}$|  
|\sqrt{f(x)|$\sqrt{f(x)}$|  
|\sum_{unteres limit}^{oberes limit}|$\sum_{unteres limit}^{oberes limit}$|  
|\lim_{x \to y} f(x)|$\lim_{x \to y} f(x)$|  
|\int_{untereslimit}^{oberes Limit} f(x) dx|$\int_{untereslimit}^{oberes Limit} f(x) dx$|  
|<hr>|<hr>|  
|\geq | $\geq$|   
|\leq|$\leq$|  
|\neq|$\neq$|   
|\in|$\in$| 
|\approx|$\approx$|
|\propto|$\propto$|
|\sim|$\sim$|
|<hr>|<hr>|    
|\vec{v}|$\vec{v}$|  
|\langle|$\langle$|  
|\rangle|$\rangle$|
|\pm|$\pm$|
|\mp|$\mp$|
|\infty|$\infty$|
|\dots|$\dots$|

Allgemeine Regeln:
---
- Sonderzeichen:  
    - Da einige zeichen in Markdown funktionen darstellen müssen diese Zeichen mit einem \\ vorangestellt werden   
- Griechisches Buchstaben:   
    - Um Griechische Buchstaben zuschreiben ein \\ und anschließend den Namen des Buchstabens:    
        - \alpha $\to \alpha$  
        - \beta  $\to \beta$  
        - \Sigma $\to \Sigma$  
        - \sigma $\to \sigma$  
        - $\vdots \to \vdots$    
 
- Zussamengesetzte Symbole:  
    - Gerade bei komplexeren Formeln kann es von nöten sein verschiedene Symbole zu kombinieren. Dafür einfach den ersten Befehl an den zweiten anknüpfen:  
          \vec{\vec{\vec{\sigma}}} = $\vec{\vec{\vec{\sigma}}}$
    - Ausserdem gibt es noch einen Befehl namens '\sideset' mit diesem befehl lassen sich Symbole in bezug zu anderen Symbolen setzen:
    - Sideset hat eine etwas komplizierte Syntax deswegen hier eine etwas ausführliche Erklärung:
      - Sideset ist ein Befehl, welcher Symbole und text in bezug zu einem anderen setzt.
      - Dabei können bis zu 6 Zeichenketten direkt über Sideset verortet werden, Zusätzliche Positionen (rechts oben und rechts unten) über Sideset zu definieren ist nicht möglich und wäre auch redundant.  
  \sideset{\_{links-unten}^{links-oben}{\_{rechts-unten}^{rechts_oben}}{das Zeichen}_{mittig-unten}^{mittig-oben}
    wenn ihr sideset benutzt aber keine symbole rechts oder links vom Symbol haben wollt müsst ihr _trotzdem_ die beiden ersten Klammern schreiben auch wenn sie leer bleiben!
hier ein paar beispiele:

|Schreibweise|Umsetzung|  
|:-----------|--------:| 
|\sideset{\_{a}^{b}{\_{c}^{d}}{\prod}\_{e}^{f}|$ \sideset{_{a}^{b}}{_{c}^{d}}{\prod}_{e}^{f} $
|\sideset{}{}{\sum}\_{e}^{f}|$ \sideset{}{}{\sum}_{e}^{f} $|
|\sideset{\sum}\_{e}^{f}|$ \sideset{\sum}_{e}^{f} $|



 
Falls ihr in einer Formel ein wenig text haben möchtet müsst ihr den Text mit \text{} markieren:
\text{eure Anmerkung}


$$
Das hier ist text ohne ohne Markierung
$$
$$
\text {Das hier ist text mit Markierung}
$$

## Casses

Desweiteren gibt es noch "cases" die können genutzt werden um Verhältnisse darzustellen:

\text{ein text auf den ihr euch bezieht...}\\begin{ cases}  
Fall 1 \\  
Fall 2 \\  
Fall 3 \\  
\cdots \\  
\end{cases}  

$$
\text{ein text auf den ihr euch bezieht...}\begin{cases}
Fall 1 \\
Fall 2 \\
Fall 3 \\
\cdots \\
\end{cases}
$$

<!--
Es gibt auch eine alternative Möglichkeit mit den Befehlen \left und \right.
Diese Befehle kennen wir von weiter oben als wir Klammern richtig skallieren wollten. 

Dabei gilt es zu wissen dass wenn wir nur \left oder nur \right setzt gibt es einen Fehler.
Deswegen geben wir immer sowohl \left als auch \right an übergeben aber dem Befehl den wir eingentlich __nicht__ wollen einen "." dadurch wird die Klammer leer gelassen.

Zum Beispiel so:

\text{ein text auf den ihr euch bezieht...} \left{  
Fall 1 \\ \\    
Fall 2 \\ \\   
Fall 3 \\ \\   
\cdots \\ \\   
\right .


$$
\begin{align*}
\text{ein text auf den ihr euch bezieht...}\left \{  
Fall 1 \\  
Fall 2 \\  
Fall 3 \\  
\cdots \\  
\right .
\end{align*}
$$
-->

in der Kombinantion sind ganz verrückte dinge möglich:

$$
\sideset{_{\sigma}^{\vec{\tau}}}{}{\aleph} \propto \oint_{0}^{\infty}\left( \frac{\underline{\alpha}}{\pm \sqrt{25}} \right) \quad \text{mit x }\in \begin{cases} [-\infty | 3]\\ [100|\infty] \end{cases} 
$$

PS: das hier gezeigte Beispiel hat keine (mir bekannte) sinnvolle Bedeutung und dient einzig dem Beispiel der Funktionen von $\LaTeX$

[Weitere Mathe Symbole hier klicken!](https://rpruim.github.io/s341/S19/from-class/MathinRmd.html)
