

<a id="Debuggen_von_Hand">
    
## Debuggen von Hand
</a>

Es kommt immer wieder vor das ihr euch auf Fehler suche in einem Programm begeben müsst und ihr aus irgendeinem grund keine Fehler meldungen bekommt (zum Beispiel in der Prüfung). Dann kann es sinnvoll sein sich "händisch" durch das Programm durchzuarbeiten.
Viele der hier vorgestellten Methoden sind schon automatiesiert in vielen Entwicklungsumgebungen vorhanden. Die Ausführungen hier gehen aber erst mal davon aus das ihr __nur__ Stift und Papier habt, und vielleicht einen Taschenrechner (wie in der Prüfung), und damit den Fehler in eurer Schleife / Funktion finden müsst.


<a id="Variablen_im_Blick">

### Variablen im Blick behalten
</a>

Viele mag es unmöglich erscheinen wie ein Computer "zu denken" eigentlich ist das aber nicht so schwer. Gerade wenn ihr einen Taschenrechner zur hand habt ist das schwierigste erst mal alle Variablen im Kopf zubehalten.
Damit uns das Leichter fellt machen wir uns Listen in denen alle Variablen aufgelistet sind und ihre Momentanen und vorhergenden Inhalte.

Eine solche Liste könnte zum Beispiel so ausseehen:

|lf.Nr. n |Variable 1|Variable 2| $\dots$|Variable n|Besonderes|
|---------|----------|----------|--------|----------|----------|
|1        | 2 | 3|$\dots$| False| | 
|2        | 3 | 6|$\dots$| True| | 
|3        | 4 | 9|$\dots$| False| |
|4        | 5 | 12 |$\dots$|True | |
|5        | 6 | 15|$\dots$| True| X |
|$\vdots$ |$\vdots$ |$\vdots$ |$\ddots$ |$\vdots$ |$\vdots$ |
|Verhalten| n+1| Variable 1 * 3|(n-1)+2|True wenn n % 2 = 0|Bei n = 5 <br>Sollte Variable n = False <br>sein ist aber True |

#### Erklärung:

die Laufende Nummer benutzen wir um einen Überblick über unseren Momentanen Zustand zu behalten.  
In die Zeile Verhalten ist steht entweder das was wir erwarten würden (also das was wir Programiert haben) oder wenn wir nicht wissen was wir erwarten schreiben wir hin wie sich die Zahlen zu einander verhalten, bzw sich zum Index verhalten.  
In der Spalte Besonderes machen wir immer dann ein Kreuz wenn sich etwas entweder Falsch verhält, oder sich eine Variable plötzlich ihr Verhalten ändert.
Auch gehört in diese Zeile wenn irgendwas im Programm passiert das ihr sonst nicht abbilden könnt, eine andere Funktion wird aufgerufen, oder der Computer geht in Flammen auf etc...


Um auf diese Tabelle zukommen ist aber nicht immer einfach gerade bei größeren Funktionen ist es oft schwierig allen Variablen in einer einzigen Tabelle zu verfolgen. Gerade bei verschachtelten for-Schleifen wird es schnell schwierig alle laufvariablen in einer Tabelle unterzubringen. Gerade dann ist es Sinnvoll mehrere Tabellen zu machen. Wie viele ist euch überlassen, die Tabellen sind nur eine bessere Darstellung um einen Überblick über die Variablen zu behalten.

Im Allgemeinen gehen wir aber wie folgt vor: 
1. Wir Analysieren unsere Funktion oder Schleife und überlegen uns was wir intuitiv erwarten würden, wenn wir uns auf etwas keinen reim machen können schreiben wir das auch so auf. Das Dient uns später um Unerwartetes besser zu finden falsche Annahmen sind hier nicht so schlimm

2. Wir erstellen uns so viele Tabellen wie wir brauchen <font color="red"><b>WICHTIG: ES MÜSSEN ALLE VARIABLEN IRGENDWO VORKOMMEN </b></font>

3. Wir gehen von oben nach unten durch: was macht das Programm gerade in dieser Zeile und NUR in dieser Zeile.<br> Wenn sich eine Variable ändert fügen wir das in die Tabelle ein wenn, wenn sich nur eine Variable ändert und die anderen unverändert bleiben müsst ihr euch ausserdem überlegen ob ihr alle Variablen nochmal mit ihren alten werten hinschreibt oder ihr die entsprechenden kästen leerlast uns später befüllt. Das müsst ihr entscheiden BEVOR ihr anfangt die Tabelle zu befüllen und das dann auch so durchziehen.

4. Wenn wir alle Felder belegt haben in unserer Tabelle befüllt haben, können wir nochmal unseren Programmcode anschauen, diesmal suchen wir aber nicht unsere Variablen sondern wir suchen das was uns das ergebniss ausgibt also z.B.: ```return```, oder ```print```.<br> Um alle ausgaben suchen wir uns in den Tabellen die entsprechenden Werte aus, und schreiben sie auf:

   - Für jede neue Printanweisung eine neue Zeile (Standartfall)
   - Die Variablen tauschen nicht ihre Position, von links nach rechts werden die Variablen einzeln ausgelesen.
   - Print schaut sich immer nur die Endwerte der Variablen an wie sie gerade übergeben werden die Position der print-Anweisung ist also wichtig für den Wert der Ausgegeben wird


#### Beispiel:

```python
ksum=0
sumzeilen=0
for i in range(0,2):
    for j in range(i,3):
        for k in range(j-i,j+1):
            print('%d %d %d' % (i,j,k))
            sumzeilen=sumzeilen+1
            ksum=ksum+k;
print ('%d %d' % (sumzeilen,ksum))
```

1. Analyse:
    - es gibt 3-Schleifen, die jeweils untere Schleife ist von der oberen abhängig
    - am in der 3. Schleife gibt es einen ```print``` befehl der die momentanen laufvariablen ausgibt
    - Es gibt 5 Variablen davon sind 3 Laufvariablen:
      - Laufvariablen:
        - i : liegt zwischen 0 und 1 (obere Grenze = 2)
        - j : liegt zwischen 0 (für i = 0) und 2 (obere Grenze = 3)
        - k : liegt zwischen 0 (für i= 0 & j=0) und 4 (j = 3)
       - Unabhängige Variablen:
         - summzeilen : fängt bei 0 an und wird immer größer
         - ksum : addiert alle k-Werte zusammen
    - am ende des Programms gibt es eine ```print``` anweisung die die endwerte von "summzeilen" und "ksum" ausgibt
 
2. Tabellen
    - Regeln zum erstellen der Tabellen:
       1. immer wenn sich i verändert mache ich eine neue Tabelle und schreibe mir den veränderten Wertebereich von j auf.
       3. immer wenn sich j verändert schreibe ich für k meinen neuen Wertebereich auf
       4. Alle meine Tabellen sind in ihrem aufbau gleich
       5. Am Anfang schreibe ich mir auf womit alle Variablen anfangen
       6. Ich schreibe statt der laufenden Nummer, auf wie viele Iterationen die Schleifen *vollendet* haben.
       7. Ich schreibe mir immer nur die Endwerte der Variablen für die momentane Iteration auf (alternative wären immer die Startwerte)
       8. Ich schreibe wenn ich auf eine Printanweisung treffe das __NICHT__ auf

3. Es gibt 2 ```print``` anweisungen, die erste ist in der k-Schleife drin und die zweite in ausserhalb aller Schleifen:

Dann kommt ihr mit einfachen Tabellen nicht mehr weiter, dann brauchen wir zu den Tabellen noch zusätzlich eine Idee was das Programm eigentlich macht. Dazu nutzen wir Flow-Charts.

#### Wie funktioniert ein Flowchart?

Ein Flowchart fängt irgendwo an und dann folgt ihr den Pfeilen zu verschiedenen Feldern. Diese Felder stellen entweder eine Abfrage, eine Funktion, eine Klasse oder eine Ausgabe dar. Auch wenn es manchmal verlockend ist und "einfach Sinn ergeben" würde, stellt ein Feld IMMER NUR EINE sache dar.

Neben der Flowchart braucht ihr immer noch eure Tabellen mit allen Variablen. In der Flowchart wird nur gezeigt was mit den Variablen passiert nicht welchen wert die Variablen haben.

Im Allgemeinen könnt ihr jedes Programm oder jeden Programmschnipsel in eine Flowchart umwandeln (und andersrum gehts auch!).
Wenn ihr euren code habt könnt ihr ihn mit [pyflowcharts](https://pypi.org/project/pyflowchart/) in automatisch in in eine Flow Chart umwandeln. Hier genauer darauf einzugehen würde zuweit gehen schaut euch dafür einfach die [Dokumentation](https://pypi.org/project/pyflowchart/) an.

#### Beispiel

Wie oben schon ausgeführt bringt uns ein Flowchart wenig wenn wir uns nicht auch die Möglichkeiten anschauen was die Variablen machen. Da es sehr aufwendig ist die Variablen "durchzurechnen" reicht es wenn wir uns besondere Bedingungen anschauen. Diese Bedingungen finden wir entweder, weil das Programm partout nicht tun will was wir wollen, oder indem wir uns überlegen was der Nutzer machen könnte womit wir nicht rechnen.  
Im folgenden schauen wir uns einfaches Beispiel an. In diesem Beispiel wird der Monat angegeben und es kommt die Anzahl der Tage raus, die der Monat hat.

```python
def motage(m):
    if m == 2:
        d = 28
    elif m in [1,3,5,7,8,10,12]:
        d=31
    elif m in [4,6,9,11]:
        d=30
    else:
        d=-1
    return d

tage=motage(4)

print(tage)

```

1. Analyse:
   - es gibt eine Funktion "motage" diese Funktion nimmt eine Variable "m" an.
   - innerhalb der Funktion "motage" gibt es die Variable "d" welche am ende zurückgeben wird
        - es gibt 1 if-Abfrage diese wird ausgeführt wenn m == 2 ist.
        - Wenn die 2 elif-Abfragen:
          - die 1. wird ausgeführt wenn m entweder 1,3,5,7,8,10 oder 12 ist
          - die 2. wird ausgeführt wenn m entweder 4, 6, 9 oder 11 ist
        - Wenn keine der Abfragen ausgeführt wird gilt d = -1
    - am Anfang der Ausführung (nicht am anfang des Programms) wir in der Variable "tage" die ausgabe der Funktion "motage" gespeichert
    - am ende des Programms wird die Variable "tage" Ausgegeben
